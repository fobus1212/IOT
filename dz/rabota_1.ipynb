```python

import numpy as np
import pandas as pd
from sklearn.metrics import mean_squared_error, r2_score


data = pd.read_csv('insurance - insurance.csv')


data = pd.get_dummies(data, columns=['sex', 'smoker'], drop_first=True) 
X = data[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']] 
y = data['charges']  


X = (X - X.mean()) / X.std()

X['intercept'] = 1


theta = np.zeros(X.shape[1])


alpha = 0.1
iterations = 1000

mse_history = []
r2_history = []


m = len(y)  

for _ in range(iterations):
  
    predictions = np.dot(X, theta)
    
  
    error = predictions - y
   
    mse = mean_squared_error(y, predictions)
    mse_history.append(mse)
    
   
    r2 = r2_score(y, predictions)
    r2_history.append(r2)
    
  
    for i in range(X.shape[1]):
        gradient = (1/m) * np.sum(error * X.iloc[:, i])
        theta[i] = theta[i] - alpha * gradient


X_test = pd.DataFrame({'age': [30], 'bmi': [25], 'children': [2], 'sex_male': [1], 'smoker_yes': [0], 'intercept': [1]})
predicted_charge = np.dot(X_test, theta)
print(f'Прогноз стоимости страховки: {predicted_charge[0]:.2f}')


for i, (mse, r2) in enumerate(zip(mse_history, r2_history)):
    print(f'Итерация {i+1}: MSE = {mse:.4f}, R-squared = {r2:.4f}')
# не совсем понимал как работает mse и как его интерпритировать , поэтому включил доп метрики r-squared как говорит интернет насколько он сильнее стримится к 1 тем  
# больше точность , тк 0.7497 на последней итерации сводится примерно к 75% не густо , но я не понимаю как увеличить точность , колличества итераций и времени обучени ни к чему 
#  не приводят , пробовал убрать из атрибутов х колличество детей , точность не изменилась , думаю задание "годно" до субботы вечера , если непрвильно или нужны корриктеровки ,  
# прошу написать в комментариях в гугл классе 
```
