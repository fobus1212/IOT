Приведены два датасета: время использования одними и теми же пользователями личного кабинета сайта до редизайна и после. Проверьте гипотезу, что после редизайна время использования личного кабинета на сайте изменилось (увеличилось или уменьшилось). Установите уровень статистической значимости 0.05.


$H_0$  после редизайна время использования личного кабинета на сайте не изменилось <br> 
$H_1$ после редизайна время использования личного кабинета на сайте изменилось.

```python
alpha = 0.05 
 
results = ss.ttest_rel( 
    time_before,   
    time_after , 
    alternative='two-sided')  # В данном случае, так как гипотеза утверждает, что время использования личного кабинета на сайте может как увеличиться, 
# так и уменьшиться после редизайна, я выбрал альтернативную гипотезу two-sided 


print(f'p-value: {results.pvalue}') 
 
if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу ')
else:
    print('Нет оснований отвергнуть нулевую гипотезу')
```
