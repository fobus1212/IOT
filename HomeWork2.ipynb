```python
for i in  Myarray :  
    with open(i, 'r', encoding='utf-8') as file:
        text = file.read()
        text = text.lower()
        tokens = word_tokenize(text)

    tagged_words = pos_tag(tokens)

    pos_counts = {'VBZ': 0, 'VBN': 0, 'VB': 0, 'VBD': 0, 'VBG': 0, 'VBP': 0, 
              'NN': 0, 'NNS': 0, 'NNP': 0, 
              'JJ': 0, 'JJR': 0, 'JJS': 0}

    for word, tag in tagged_words:
        if tag in pos_counts:
            pos_counts[tag] += 1 
 
 
    print(i)
    print("Глаголы:")
    for pos in ['VBZ', 'VBN', 'VB', 'VBD', 'VBG', 'VBP']:
        print(f'{pos}: {pos_counts[pos]}')

    print("\nСуществительные:")
    for pos in ['NN', 'NNS', 'NNP']:
        print(f'{pos}: {pos_counts[pos]}')

    print("\nПрилагательные:")
    for pos in ['JJ', 'JJR', 'JJS']:
        print(f'{pos}: {pos_counts[pos]}') 
         
    total_words = sum(pos_counts.values())
    print(f"\nОбщее количество слов: {total_words}")


    noun_percentage = (pos_counts.get('NN', 0) + pos_counts.get('NNS', 0) + pos_counts.get('NNP', 0)) / total_words * 100
    adj_percentage = (pos_counts.get('JJ', 0) + pos_counts.get('JJR', 0) + pos_counts.get('JJS', 0)) / total_words * 100
    verb_percentage = (pos_counts.get('VBZ', 0) + pos_counts.get('VBN', 0) + pos_counts.get('VB', 0) + pos_counts.get('VBD', 0) + pos_counts.get('VBG', 0) + pos_counts.get('VBP', 0)) / total_words * 100

    print(f"\nПроцентное соотношение существительных: {noun_percentage:.2f}%")
    print(f"Процентное соотношение прилагательных: {adj_percentage:.2f}%")
    print(f"Процентное соотношение глаголов: {verb_percentage:.2f}%")
    
```

Глаголы:
