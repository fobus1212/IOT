```python
import nltk
from nltk.parse.corenlp import CoreNLPServer
from nltk.parse import CoreNLPParser


server = CoreNLPServer("path/to/corenlp/stanford-corenlp-4.2.0.jar", "path/to/corenlp/stanford-corenlp-4.2.0-models.jar")
server.start()


nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('treebank')


parser = CoreNLPParser(url='http://localhost:9000')


with open("C:/Users/fobus/Desktop/GIT_TEST/nltk/HomeWork2/The Hobbit 1937.txt", "r", encoding="utf-8") as file:
    text = file.read()


tokens = nltk.word_tokenize(text)


parsed_sentences = list(parser.parse_sents(tokens[:100]))


for sentence in parsed_sentences:
    for tree in sentence:
        tree.pretty_print()

server.stop()
```
