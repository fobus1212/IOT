```python
from gensim.models.fasttext import FastText as GensimFastText
import gensim.downloader as api

model = api.load('fasttext-wiki-news-subwords-300')

with open('C:/Users/fobus/Desktop/GIT_TEST/nltk/HomeWork2/The Children 2007.txt', 'r', encoding='utf-8') as file:
    text = file.read()


words = text.split()[:100]

similar_words = []
for word in words:
    if word in model:
       
        most_similar = model.most_similar(word, topn=5)
        similar_words.append((word, most_similar))


for word, similar in similar_words:
    print(f'Слово: {word}')
    for sim_word, score in similar:
        print(f'Похожее слово: {sim_word}, Оценка близости: {score}')

```
